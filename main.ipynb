The second approach is topological sorting. This sorting method is similar to a DFS except this time we modify it by using a temporary stack. In topological sorting, we don't print the vertex immediately, we recursively call topological sorting for all its adjacent vertices then push it to a stack. Finally we print the contents of the stack. Note that a vertex is pushed to the stack only when all of its adjavent vertices are already in stack. 
The complexity of topological sorting is O(V+E)
